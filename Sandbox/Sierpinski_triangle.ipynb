# Sierpinski triangle plotting

```python
from random import choice
from matplotlib import pyplot as plt
```

```python
def t_1(p):
    x = p[0]
    y = p[1]
    x1 = 0.5 * x
    y1 = 0.5 * y
    return x1,y1

def t_2(p):
    x = p[0]
    y = p[1]
    x1 = 0.5 * x + 0.5
    y1 = 0.5 * y + 0.5
    return x1,y1

def t_3(p):
    x = p[0]
    y = p[1]
    x1 = 0.5 * x + 1.0
    y1 = 0.5 * y
    return x1,y1
```

```python
transformations = [t_1, t_2, t_3]
a1 = [0]
b1 = [0]
a,b = 0,0
```

```python
for i in range(999999):
    trans = choice(transformations)
    a,b = trans((a,b))
    a1.append(a)
    b1.append(b)
```

```python
plt.rc('figure', figsize=(16,16))
plt.plot(a1,b1, 'o')
```
