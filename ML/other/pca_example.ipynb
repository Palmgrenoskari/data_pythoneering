# Principal Component Analysis example

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

##### Load the dataset

```python
data = pd.read_csv('../data/California_Real_Estate.csv', sep= ';')
df = data.copy()
df
```

##### Discarding unsold properties

* Status = 0

```python
df_sold = df[df['Status'] == 1]
df_sold
```

##### Standardizing

* -> Mean 0 , STDEV 1

```python
scaler = StandardScaler()
df_sold_scaled = scaler.fit_transform(df_sold)
```

```python
df_sold_scaled
```

##### PCA

```python
# 8 components, since we have 8 features and didnt specify n_components = _
pca = PCA()
pca.fit_transform(df_sold_scaled)
```

```python
# Returns how much each component explains of the total variance. 
pca.explained_variance_ratio_
```

##### Barchart - How much each component explains of the total variance

```python
plt.figure(figsize = (11,6))
components = ['Component 1', 'Component 2', 'Component 3', 'Component 4', 'Component 5', 'Component 6', 'Component 7', 'Component 8']
explained_variances = pca.explained_variance_ratio_

plt.bar(components, explained_variances)
plt.title('Explained variance by Principal Components')
plt.xlabel('Components')
plt.ylabel('Explained Variance')
plt.show()
```

##### Cumulative explained variance until a certain component - Chart

```python
plt.figure(figsize=(11,6))
plt.plot(range(1,9),
         pca.explained_variance_ratio_.cumsum(),
         marker='o',
         linestyle = '--')
plt.title('Explained variance by Components')
plt.xlabel('Num of Components')
plt.ylabel('Cumulative explained variance')
plt.show()
```

* Good rule of thumb is to retain at least 80% of the total variance.
* -> 4 components looks like a solid num of components

##### PCA Analysis and interpretation

```python
# 4 components
pca = PCA(n_components=4)
```

```python
pca.fit(df_sold_scaled)
```

```python
# Correlation of each of our variable with the corresponding component
# We have 4 components and 8 features -> shape = (4,8)
pca.components_
```

```python
# DF for easier interpretation
df_pca_components = pd.DataFrame(data = pca.components_,
                                 columns=df.columns.values,
                                 index=['Component 1', 'Component 2', 'Component 3', 'Component 4'])
df_pca_components
```

```python
# Heatmap for further interpretation
# Correlation values go from -1 to 1
# -1 meaning perfect negative correlation and 1 perfect positive correlation
sns.heatmap(df_pca_components,
            vmin=-1,
            vmax=1,
            cmap='RdBu',
            annot=True)

plt.yticks([0,1,2,3],
           ['Component 1', 'Component 2', 'Component 3', 'Component 4'],
           rotation = -45,
           fontsize = 9)

plt.show()
```

# Explaining our components by variables

* Component 1: Charasteristics of the property
* Component 2: Building's charasteristics
* Component 3: Seasonality of sales?
* Component 4: Type of property, but really a mixture of all features imo.
